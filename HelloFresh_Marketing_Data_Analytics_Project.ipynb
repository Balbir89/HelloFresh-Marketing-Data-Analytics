### A/B Test Simulation (Example)

Simulate an A/B test and run a t-test to demonstrate lift analysis.


```python
import numpy as np
from scipy.stats import ttest_ind

# Simulate A/B test results
group_a = np.random.normal(loc=45, scale=3, size=100)
group_b = np.random.normal(loc=43, scale=3, size=100)

# Perform t-test
t_stat, p_value = ttest_ind(group_a, group_b)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Statistically significant difference between A and B")
else:
    print("❌ No significant difference between A and B")
```

### Interactive Visualization: Channel Efficiency (ROAS) Over Time

This chart shows the return on ad spend (ROAS) by marketing channel and year — how much revenue HelloFresh earns for each euro spent.


```python
import plotly.express as px

# Calculate Marketing Spend (€M)
df['Marketing Spend (€M)'] = df['Revenue (€M)'] * (df['Mkt Spend (%Rev)'] / 100)

# Calculate ROAS
df['ROAS'] = df['Revenue (€M)'] / df['Marketing Spend (€M)']

# Aggregate average ROAS by Year and Channel
roas_channel = df.groupby(['Year', 'Channel'])['ROAS'].mean().reset_index()

# Plot interactive line chart of ROAS by Channel
fig = px.line(
    roas_channel,
    x='Year',
    y='ROAS',
    color='Channel',
    markers=True,
    title='Channel Efficiency: ROAS Over Time'
)

fig.show()
```

### Seasonal Decomposition of Revenue

Decomposing the revenue time series into trend, seasonal, and residual components.


```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd

# Assuming you have a 'Date' column or monthly data; if not, this is an example with yearly data only.
# Example: Convert Year to datetime (using January 1st)
revenue_year['Date'] = pd.to_datetime(revenue_year['Year'], format='%Y')

# Set Date as index
revenue_year.set_index('Date', inplace=True)

# Perform seasonal decomposition - model='additive' or 'multiplicative'
result = seasonal_decompose(revenue_year['Revenue (€M)'], model='additive', period=1) # period=1 for yearly data

# Plot decomposition
result.plot()
plt.show()
```

### Forecasting HelloFresh Revenue Using Prophet

Forecast future revenue based on historical trends using the Prophet model, which handles seasonality and trends automatically.


```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data for Prophet: must have columns 'ds' (date) and 'y' (value)
df_prophet = revenue_year.reset_index()[['Date', 'Revenue (€M)']]
df_prophet.rename(columns={'Date': 'ds', 'Revenue (€M)': 'y'}, inplace=True)

# Initialize and fit the model
model = Prophet(yearly_seasonality=True)
model.fit(df_prophet)

# Create dataframe to hold predictions for next 3 years (12 quarters or 36 months if monthly data)
future = model.make_future_dataframe(periods=3, freq='Y')  # 3 years yearly forecast

# Predict future revenue
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title('HelloFresh Revenue Forecast (Next 3 Years)')
plt.show()
```

### Interpretation of Revenue Forecast

The Prophet forecast for HelloFresh’s revenue over the next three years indicates a continued upward growth trend, suggesting strong market demand and effective marketing strategies to date. The model projects steady increases in revenue year-over-year, which supports maintaining or cautiously increasing marketing investments to capitalize on growth momentum.

The confidence intervals (shaded areas) widen gradually, reflecting increased uncertainty further into the future, which is typical for forecasting models. This highlights the importance of regularly updating forecasts with new data.

While this model uses historical revenue alone, incorporating additional variables such as marketing spend, conversion rates, or external market factors could enhance forecast accuracy and provide deeper strategic insights.

# Prepare data with an additional regressor

```python
df_prophet['mkt_spend'] = df.groupby('Year')['Marketing Spend (€M)'].sum().values

model = Prophet(yearly_seasonality=True)
model.add_regressor('mkt_spend')

model.fit(df_prophet)

future = model.make_future_dataframe(periods=3, freq='Y')
# For future mkt_spend, you can either fill with recent average or estimate
future['mkt_spend'] = df_prophet['mkt_spend'].mean()

forecast = model.predict(future)
fig2 = model.plot(forecast)
plt.title('Revenue Forecast with Marketing Spend as Regressor')
plt.show()
```
